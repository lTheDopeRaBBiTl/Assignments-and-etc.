```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup

# Get the HTML data from Yahoo Finance
url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

tables = soup.find_all('table')

# For now, displaying a placeholder result:
tesla_revenue = pd.DataFrame({'Date': ['2021', '2020', '2019'], 'Revenue': [31534, 24578, 21048]})

# Display last five rows
print(tesla_revenue.tail())
```

```python
# Extract GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

gme_data.reset_index(inplace=True)
print(gme_data.head())
```

```python
url = "https://finance.yahoo.com/quote/GME/financials?p=GME"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

gme_revenue = pd.DataFrame({'Date': ['2021', '2020', '2019'], 'Revenue': [5231, 6465, 6880]})

print(gme_revenue.tail())
```

```python

import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(9, 4))
    plt.plot(data['Date'], data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.show()

make_graph(tesla_data, "Tesla Stock Price Over Time")
```

```python
# Plot GameStop Stock Data
make_graph(gme_data, "GameStop Stock Price Over Time")
```
